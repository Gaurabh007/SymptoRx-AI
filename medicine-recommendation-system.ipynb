## => Split the Data in Train-Test Split

```python
X = df.drop('prognosis' ,axis=1)
y = df['prognosis']
```

```python
X.head(5)
```

```python
y.head(5)
```

#### Now we need to labelEncode our **y data** because the machine learn binary language fast and more accurate so here we use labelEncoder from scikit-learn library to transform the data

```python
le = LabelEncoder()
le.fit(y)
Y = le.transform(y)
```

```python
Y # here 15 means Fungal infection and 27 means impetigo and soo on..
```

```python
# split the data 
X_test,X_train,y_test,y_train = train_test_split(X,Y,test_size =0.2 , random_state=42)
```

```python
X_test.shape,X_train.shape,y_test.shape,y_train.shape
```

## => Train Top Models

```python
# Here we can train the top best models which can give best accuracy So i can create a Dictionary of models to check which model is more accurate.
models = {
      'SVC' : SVC(kernel='linear'),
      'Random Forest' :RandomForestClassifier(random_state=42,n_estimators=100),
      'KNeighbors' : KNeighborsClassifier(n_neighbors=5),
      'Gradient Boosting': GradientBoostingClassifier(random_state=42,n_estimators=100),
      'MultinomialNB' :MultinomialNB() 
}
for model_name , model in models.items():
    # Train model
    model.fit(X_train,y_train)
    # test model
    predictions = model.predict(X_test)
    # calculate accuracy
    accuracy = accuracy_score(y_test,predictions)
    # calculate confusion matrix
    cm = confusion_matrix(y_test,predictions)
    # print results
    print(f"{model_name} accuracy : {accuracy}")
    print(f"{model_name} confusion matrix :")
    print(np.array2string(cm,separator=', '))
```

## => Make predictions on single model

### On above cell all model gives the best 100% accuracy so we dont need to go with all models (you can) So i go with **SVC** model

```python
from sklearn.svm import SVC
svc = SVC(probability=True, random_state=42)  # <<== note this!
svc.fit(X_train, y_train)

y_pred = svc.predict(X_test)
acc = accuracy_score(y_pred,y_test)
acc
```

## => Recommended System and Predictions

```python
def predict_disease(user_symptoms, all_symptoms, model, le, min_symptoms=5, min_prob=0.0):
    if len(user_symptoms) < min_symptoms:
        return "Insufficient symptoms provided for a reliable prediction."
    user_input = np.zeros((1, len(all_symptoms)), dtype=int)
    for idx, s in enumerate(all_symptoms):
        if s in user_symptoms:
            user_input[0, idx] = 1
    probs = model.predict_proba(user_input)[0]
    top_proba = np.max(probs)
    best_class = np.argmax(probs)
    if top_proba < min_prob:
        return "Prediction uncertain/unreliable. Please consult a doctor."
    disease = le.inverse_transform([best_class])[0]
    return f"Predicted disease: {disease} (confidence {top_proba:.2f})"


```

```python
# Suppose user entered three symptoms
user_symptoms = ['fever', 'cough', 'sore_throat', 'headache', 'chills', 'muscle_pain']

# All possible symptoms/features in the training set
all_symptoms = X.columns.tolist()  # Or, keep a master list

# Example use (assuming your trained SVC is called 'svc_pipe'):
predict_disease(user_symptoms, all_symptoms, svc, le)
```
