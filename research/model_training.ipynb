```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
pd.set_option('display.max_columns', None)
```

```python
train_df = pd.read_csv('data\disease_prediction_dataset\Training.csv')
# med_df1 = pd.read_csv('data\medicine dataset\medicines.csv')
# med_df2 = pd.read_csv('data\medicine dataset\medical_question_answer_dataset_50000.csv')
```

## Spliting the dataset into Train and Test

```python
X = train_df.drop(columns=['prognosis'])
y = train_df['prognosis']

# Encode target text → integer labels
le = LabelEncoder()
y_enc = le.fit_transform(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y_enc, test_size=0.20, random_state=42, stratify=y_enc
)

print("Train size :", X_train.shape, "  Test size :", X_test.shape)
```

## Building The Model

### Random Forest Model

```python
rf = RandomForestClassifier(
    n_estimators=300,
    min_samples_leaf=2,
    random_state=42,
    n_jobs=-1
)
rf.fit(X_train, y_train)
```

```python
y_pred = rf.predict(X_test)

# Basic accuracy
print("\nAccuracy :", accuracy_score(y_test, y_pred))

# Get unique labels in y_test and y_pred (to handle missing classes)
unique_labels = np.unique(np.concatenate([y_test, y_pred]))

# Subset target_names to match only these unique labels
subset_names = le.inverse_transform(unique_labels)

# Classification report for present classes only
print("\nClassification Report (for present classes only):")
print(classification_report(
    y_test, y_pred,
    labels=unique_labels,          
    target_names=subset_names,     
    zero_division=0               
))
```

```python
from sklearn.model_selection import cross_val_score
cv_scores = cross_val_score(rf, X, y_enc, cv=5, n_jobs=-1)
print("CV scores:", cv_scores, "Mean:", cv_scores.mean())
```

# Our Model

```python
import pickle
model = pickle.load(open('models/randomforest.pkl','rb'))
```

```python
def predict_disease(user_symptoms, all_symptoms, model, le, min_symptoms=4):
    if len(user_symptoms) < min_symptoms:
        return "Insufficient symptoms provided for a reliable prediction."
    user_input = np.zeros((1, len(all_symptoms)), dtype=int)
    for idx, s in enumerate(all_symptoms):
        if s in user_symptoms:
            user_input[0, idx] = 1
    best_class = model.predict(user_input)[0]
    disease = le.inverse_transform([best_class])[0]
    return f"Predicted disease: {disease}"
```

```python
# Suppose user entered three symptoms
user_symptoms =['fever', 'cough', 'sore throat', 'headache', 'chills', 'muscle pain']

# All possible symptoms/features in the training set
all_symptoms = X.columns.tolist()  # Or, keep a master list

predict_disease(user_symptoms,all_symptoms,model,le)
```
